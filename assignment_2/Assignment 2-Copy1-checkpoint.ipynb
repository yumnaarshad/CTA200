```python
# improved code 2
import numpy as np 
import cmath
import matplotlib.pyplot as plt

n = 100 # number of iterations
coord = 1000

x = np.linspace(-2,2,coord)
y = np.linspace(-2,2,coord)

points = []
for num_x in x:
    for num_y in y: 
        point = (num_x,num_y)
        points.append(point)

        
# Checking whether the modulus of z values associated to each point are bounded or not 
# define bounded as being within the square domain of (-2,2)x(-2,2)
# create dictionary with 2 keys: bound, diverge and the values are a list of the associated points within the complex plane c
points_dict = {'bound':[], 'diverge':[]}
iteration_num =[]

for point in points:
    # initialize z0 = 0 
    z = 0
    #c = point[0] + 1j*point[1]]
    c = complex(point[0],point[1])
    # iterate n times to determine whether it converges or diverges
    itn = 0 #track iteration number
    diverge = False
    while (abs(z.real) <= 2) and (abs(z.imag) <= 2) and (itn <= n):
        z = z**2 + c
        itn += 1
        
    if itn < n:
        points_dict['diverge'].append(point)
        iteration_num.append(itn)
    else:
        points_dict['bound'].append(point)
        
    
num_diverge = len(points_dict['diverge'])
num_bound = len(points)-num_diverge

x_bound = np.zeros(num_bound)
y_bound = np.zeros(num_bound)
i = 0
for point in points_dict['bound']:
    x_bound[i] = point[0]
    y_bound[i] = point[1]
    i += 1
    
x_diverge = np.zeros(num_diverge)
y_diverge = np.zeros(num_diverge)
i = 0
for point in points_dict['diverge']:
    x_diverge[i] = point[0]
    y_diverge[i] = point[1]
    i += 1
    
    
### Plot diverging vs bounded (2 diff colors)
# Plot diverging points in one color
plt.plot(x_diverge, y_diverge,  linestyle = 'None', marker = 'o', markersize = 1.5, markerfacecolor = 'blue', label = 'diverging points')
# Plot bounded points in another color
plt.plot(x_bound,y_bound, linestyle = 'None', marker = 'o', markersize = 1.5, markerfacecolor = 'red', label = 'bounded points')
plt.legend()
plt.grid()
plt.xlabel('Real Axis')
plt.ylabel('Imaginary Axis')
plt.title("Assignment 2 Q1 Plot")
plt.show()
plt.close()
```

## Q2

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp


N = 1000
num = 3 # number of different gamma and beta values 

# gamma = 1/D where D is days person remains infectious
# beta = average # of contacts per person per day 
gamma_arr = [1/20, 1/2, 1/40]
beta_arr = [0.3, 0.8, 1.5]

def SIR_model(t,y,beta,gamma):
    
    S, I, R = y
    dS_dt = -beta/N * S* I 
    dI_dt = beta/N * S * I - gamma * I
    dR_dt = gamma * I 

    return [dS_dt, dI_dt, dR_dt]



fig, ax = plt.subplots(1, 3, dpi = 700, figsize = (18,6))

for i in range(num):
  
    soln = solve_ivp(fun = lambda t,y : SIR_model(t,y, beta_arr[i], gamma_arr[i]) , t_span = (0,200), y0 = [999,1,0], t_eval = np.linspace(0,200,1000))
    
    #plot solution 
    ax[i].plot(soln.t, soln.y[0],color = 'blue', label= ' S(t)')
    ax[i].plot(soln.t, soln.y[1], color = 'green',label= ' I(t)')
    ax[i].plot(soln.t, soln.y[2], color = 'red', label= ' R(t)')
    ax[i].set_title('SIR Model for: gamma='+str(gamma_arr[i])+ ' & beta='+str(beta_arr[i]))
    ax[i].set_xlabel('Time')
    ax[i].set_ylabel('Number of Individuals')
    ax[i].grid()
    ax[i].legend()

    

```
